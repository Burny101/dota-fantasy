```python
import pandas as pd
import numpy as np
import os
import json
import plotly.express as px
```

```python
with open('data/curated/player_data.json', 'r') as f:
    player_data = json.load(f)
```

```python
df = pd.DataFrame(player_data)
```

```python
df = df.astype({
    'match_id': 'string',
    'account_id': 'string',
    'hero_id': 'string',
    'radiant_team_id': 'string',
    'dire_team_id': 'string',
    'win': 'boolean',
    'firstblood_claimed': 'boolean',
    'team_id': 'string',
    'enemy_team_id': 'string',
    'valveId': 'string',
})
```
