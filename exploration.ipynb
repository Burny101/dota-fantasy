```python
def read_match_ids_from_tournament_json(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
    match_ids = [entry['match_id'] for entry in data]
    df = pd.DataFrame(match_ids, columns=['match_id'])
    return df
```

```python
def get_match_info(df_tournament):
    api_url = 'https://api.opendota.com/api/matches/'
    df_matches = []
    
    # Initialize tqdm without context manager
    pbar = tqdm(total=len(df_tournament))
    
    for _, row in df_tournament.iterrows():
        match_api_url = row['match_id']
        response = requests.get(api_url + str(match_api_url))
        time.sleep(1)

        if response.status_code == 200:
            match_data = response.json()
            df_matches.append(pd.DataFrame([match_data]))
        
        pbar.update(1)
    
    pbar.close()  
    return pd.concat(df_matches, ignore_index=True) if df_matches else pd.DataFrame()
```
